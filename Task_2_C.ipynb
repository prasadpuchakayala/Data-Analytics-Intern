# Data visualization
# Pairplot for exploring relationships between numerical variables

```python
sns.pairplot(df[['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']])
plt.show()
```

# Exploratory Data Analysis (EDA)

```python
# Distribution Analysis
plt.figure(figsize=(10, 6))
sns.histplot(df['calories'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

# Correlation Analysis

```python
# Select only numeric columns for correlation analysis
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
corr = df[numeric_cols].corr()

# Correlation Analysis
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

# Categorical Variables Analysis

```python

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='type')
plt.title('Count of Cereal Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

#Count of Cereal Manufacturers

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='mfr')
plt.title('Count of Cereal Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.show()
```

## Comparison of different types of food items


```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='type', y='calories')
plt.title('Comparison of Calories in Different Types of Food Items')
plt.xlabel('Type')
plt.ylabel('Calories')
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='type', y='sugars')
plt.title('Comparison of Sugar Content in Different Types of Food Items')
plt.xlabel('Type')
plt.ylabel('Sugar (g)')
plt.show()
```
